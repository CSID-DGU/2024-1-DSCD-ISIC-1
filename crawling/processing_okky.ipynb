```python
import pandas as pd
```

```python
okky = pd.read_csv('okkyjobs.csv')
okky
```

```python
# 각 요소 앞뒤로 따옴표를 제거
okky['skills'] = okky['skills'].str.replace("[", "")
okky['skills'] = okky['skills'].str.replace("]", "")
okky['skills'] = okky['skills'].str.replace("'", "")
okky
```

```python
unique_fields = okky['field'].unique()
unique_fields
```

```python
for i in unique_fields:
    print(i)
```

```python
def replace_field(row):
    if '풀스택개발' in row:
        return '풀스택개발'
    elif '백엔드개발' in row and '프론트엔드개발' in row:
        return '풀스택개발'
    else:
        return row

okky['field'] = okky['field'].apply(replace_field)

# 결과 출력
print(okky['field'].unique())
```

```python
def replace_field(row):
    if row == '프론트엔드개발':
        return '프론트엔드 개발자'
    elif row == '풀스택개발':
        return '풀스택 개발자'
    else:
        return row

okky['field'] = okky['field'].apply(replace_field)

# 결과 출력
print(okky['field'].unique())
```
