```python
import pandas as pd

df = pd.read_csv('../data/rallit.csv')
df = df.dropna(subset=['직업'])
df.info()  # Add parentheses here to call the function
```

```python
import numpy as np

# 조건에 따라 각 직업 유형 정의
conditions = [
    df['직업'].str.contains('풀|Full|full|웹|앱|Web|App|Software|소프트웨어', na=False),
    df['직업'].str.contains('데이터 엔지니어|Data Engineer|data engineer', na=False),
    df['직업'].str.contains('백|벡|Back|back|Node.js|서버|Server|java|Java|spring|Spring|Cloud|Reliability', na=False),
    df['직업'].str.contains('데이터|Data|data|AI|머신러닝', na=False),
    df['직업'].str.contains('데브옵스|Devops|DevOps|devops', na=False),
    df['직업'].str.contains('PM|기획|서비스|Product', na=False),
    df['직업'].str.contains('프론트|프런트|Front|front|React|react|Android|안드로이드|iOS|ios|플러터|PHP', na=False),
    df['직업'].str.contains('Sales|영업|마케터', na=False)
]

# 각 직업 유형에 해당하는 값 정의
values = [
    '풀스택 개발자',
    '데이터 엔지니어',
    '백엔드 개발자',
    '데이터 분석가',
    '데브옵스 엔지니어',
    'PM',
    '프론트엔드 개발자',
    'Sales'
]

# 조건에 따라 새로운 열에 해당하는 값을 할당
df['직무'] = np.select(conditions, values, default='기타')

df
```

```python
df[(df['직무'] == '기타') & (~df['직업'].str.contains('학생', na=False))].count()
```

```python
df = df[(df['직무'] != '기타') & (~df['직무'].isna())]
```
